```python
import math
import random


number_of_inputs = int(input("Enter the number of inputs: "))
number_of_hidden_neurons = int(input("Enter the number of hidden neurons: "))
activation_function = input("Enter the activation function (sigmoid, relu): ") # provide options: sigmoid, relu in input



print(f"Number of inputs: {number_of_inputs}")
print(f"Number of hidden: {number_of_hidden_neurons}")
print(f"Activation function: {activation_function}")


```

```python

inputs = [random.uniform(-1, 1) for _ in range(number_of_inputs)]

print(f"Inputs: {inputs}")

hidden_weights = [[random.uniform(-1, 1) for _ in range(number_of_inputs)] for _ in range(number_of_hidden_neurons)]

hidden_biases = [random.uniform(-1, 1) for _ in range(number_of_hidden_neurons)]

print(f"Hidden weights: {hidden_weights}")

if activation_function == "sigmoid":
    hidden_outputs = [1 / (1 + math.exp(-sum(inputs[i] * hidden_weights[j][i] for i in range(number_of_inputs)) + hidden_biases[j])) for j in range(number_of_hidden_neurons)]
elif activation_function == "relu":
    hidden_outputs = [max(0, sum(inputs[i] * hidden_weights[j][i] for i in range(number_of_inputs)) + hidden_biases[j]) for j in range(number_of_hidden_neurons)]


print(f"Hidden outputs: {hidden_outputs}")

output_weights = [random.uniform(-1, 1) for _ in range(number_of_hidden_neurons)]

output_bias = random.uniform(-1, 1)

print(f"Output weights: {output_weights}")
print(f"Output bias: {output_bias}")

output = sum(hidden_outputs[i] * output_weights[i] for i in range(number_of_hidden_neurons)) + output_bias

```
